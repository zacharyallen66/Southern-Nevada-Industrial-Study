```python
def create_circular_bar_chart(df, year, variable):
    # Filter data for the specified year and sort by the variable of interest
    df_year = df[df['Year'] == year].copy()
    df_year = df_year.sort_values(by=variable, ascending=True)
    
    # Initialize the figure
    fig, ax = plt.subplots(figsize=(10,10), subplot_kw=dict(polar=True))
    plt.axis('off')

    # Constants
    upperLimit = 100
    lowerLimit = 30
    labelPadding = 4

    # Compute max and min in the dataset
    max_val = df_year[variable].max()

    # Compute heights
    slope = (max_val - lowerLimit) / max_val
    heights = slope * df_year[variable] + lowerLimit

    # Compute the width of each bar
    width = 2*np.pi / len(df_year.index)

    # Compute the angle each bar is centered on
    indexes = list(range(1, len(df_year.index)+1))
    angles = [element * width for element in indexes]

    # Create a color palette
    commodities = df['Commodity_y'].unique()
    colors = sns.color_palette('hls', len(commodities))
    color_dict = {commodity: color for commodity, color in zip(commodities, colors)}

    # Draw bars
    bars = ax.bar(
        x=angles, 
        height=heights, 
        width=width, 
        bottom=0,  # Start bars from the center
        color=[color_dict[commodity] for commodity in df_year['Commodity_y']],
        edgecolor="white",
        linewidth=2,
    )

    # Add labels
    for bar, angle, height, label in zip(bars, angles, heights, df_year['Commodity_y']):
        rotation = np.rad2deg(angle)
        alignment = ""
        if angle >= np.pi/2 and angle < 3*np.pi/2:
            alignment = "right"
            rotation = rotation + 180
        else: 
            alignment = "left"

        ax.text(
            x=angle, 
            y=bar.get_height() + labelPadding, 
            s=label, 
            ha=alignment, 
            va='center', 
            rotation=rotation, 
            rotation_mode="anchor"
        )

    # Create a legend
    ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
    ax2.axis('off')
    legend_elements = [Patch(facecolor=color_dict[commodity], label=commodity) for commodity in commodities]
    ax2.legend(handles=legend_elements, loc='upper left', bbox_to_anchor=(1.05, 1))

    plt.title(f'{variable} Summary for {year}')
    plt.show()
```

```python
# # Creating separate dataframes for tons and value where their values are greater than 0
tons_summary = summary[summary['Tons'] > 0]
value_summary = summary[summary['Value'] > 0]

years = summary['Year'].unique()

# Run the function for each year for tons
for year in years:
    create_circular_bar_chart(tons_summary, year, 'Tons')

# Run the function for each year for value
for year in years:
    create_circular_bar_chart(value_summary, year, 'Value')
```
